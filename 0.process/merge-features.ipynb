```python
import os
import pandas as pd
```

```python
file = os.path.join("data", "training_feature_select_robust.csv")
eb_train_df = pd.read_csv(file, index_col=0)

print(eb_train_df.shape)
eb_train_df.head()
```

```python
file = os.path.join("data", "cnn_train_feature_select_robustize.csv")
cnn_train_df = pd.read_csv(file, index_col=0)

print(cnn_train_df.shape)
cnn_train_df.head()
```

```python
full_train_df = (
    eb_train_df.merge(
        cnn_train_df,
        left_index=True,
        right_index=True
    )
)

file = os.path.join("data", "merged_training.csv")
full_train_df.to_csv(file, sep=',')

print(full_train_df.shape)
full_train_df.head()
```

```python
file = os.path.join("data", "validation_feature_select_robust.csv")
eb_val_df = pd.read_csv(file, index_col=1)

print(eb_val_df.shape)
eb_val_df.head()
```

```python
file = os.path.join("data", "cnn_validation_feature_select_robustize.csv")
cnn_val_df = pd.read_csv(file, index_col=0)

print(cnn_val_df.shape)
cnn_val_df.head()
```

```python
full_validation_df = (
    eb_val_df.merge(
        cnn_val_df,
        left_index=True,
        right_index=True
    )
)

file = os.path.join("data", "merged_validation.csv")
full_validation_df.to_csv(file, sep=',')

print(full_validation_df.shape)
full_validation_df.head()
```
