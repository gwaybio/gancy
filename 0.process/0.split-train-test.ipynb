# Obtain Training and Testing Set from Training Data

```python
import os
import pandas as pd
```

```python
# These were randomly selected
file = os.path.join("data", "test_samples.csv")
test_samples = pd.read_csv(file)
test_samples.head()
```

```python
file = os.path.join("data", "training_data.csv")
df = pd.read_csv(file)

print(df.shape)
df.head()
```

## Split Off True Test Set and True Training Set

```python
train_df = df.query("cell_code not in @test_samples.cell_code")
test_df = df.query("cell_code in @test_samples.cell_code")
```

```python
assert len(set(train_df.cell_code).intersection(set(test_df.cell_code))) == 0
```

```python
file = os.path.join("data", "train_ebfeatures.csv")
train_df.to_csv(file, sep=',', index=False)
```

```python
file = os.path.join("data", "test_ebfeatures.csv")
test_df.to_csv(file, sep=',', index=False)
```
